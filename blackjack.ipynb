```python
cards=[11,2,3,4,5,6,7,8,9,10,10,10,10]
game=1
while game>0:
    ask=input("do you want to play?'y/n': ").lower()

    if ask=="y":
        player=[]
        computer=[]
        import random
        p1=random.choice(cards)
        p2=random.choice(cards)
        player.extend([p1,p2])
        c1=random.choice(cards)
        c2=random.choice(cards)
        computer.extend([c1,c2])
        print(f"your cards: {player},current score: {sum(player)}")
        print(f"computer's first card: {computer[0]}")

        ask1=input("type'y' to get another card,type 'n' to pass")
        if ask1=="y":
            p3=random.choice(cards)
            if (p3==11) and (21-sum(player)<3):
                p3=1

            player.append(p3)
            final_player_sum=sum(player)
        elif ask1=="n":
            player.append(0)
            final_player_sum=sum(player)
        if sum(computer)<17:
            c3=random.choice(cards)
            computer.append(c3)
            final_comp_sum=sum(computer)
        else:
            final_comp_sum=sum(computer)


        print(f"your final hand {player} ,final score: {final_player_sum}")
        print(f"Computer's final hand: {computer}, final score: {final_comp_sum}")

        if (final_player_sum<21) and (final_comp_sum<21):
            if final_comp_sum <final_player_sum :
                print("you win")
            elif final_comp_sum ==final_player_sum :
                print("match draw")
            elif final_comp_sum >final_player_sum:
                print("you lost")

        elif final_comp_sum>21:
            print("you win,computer busted")
        elif final_player_sum>21:
            print("you lost,you busted")



    if ask=="n":
        game=0


    
    
```
